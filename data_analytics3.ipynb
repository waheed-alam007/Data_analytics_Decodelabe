```python
import pandas as pd
import sqlite3      # For using SQL inside Python
```

```python
df = pd.read_excel("Cleaned_Dataset.xlsx")
df.head()
```

```python
# STEP 3:Create an in-memory SQL database
conn = sqlite3.connect(":memory:")
df.to_sql("orders", conn, index=False, if_exists='replace')   #Store dataset into SQL table named 'orders'
```

```python
#  STEP 4:  SELECT query (view first 5 rows)
print("\n===== ALL DATA =====")
query1 = "SELECT * FROM orders LIMIT 5"
print(pd.read_sql(query1, conn))
```

```python
#  STEP 5: WHERE CLAUSE
# Find high-value orders (TotalPrice > 1000)
print("\n===== HIGH VALUE ORDERS =====")
query2 = """
SELECT OrderID, TotalPrice
FROM orders
WHERE TotalPrice > 1000
"""
print(pd.read_sql(query2, conn))
```

```python
# STEP 6: ORDER BY (Sorting Data)
#  Get top 5 highest orders
print("\n===== TOP 5 EXPENSIVE ORDERS =====")
query3 = """
SELECT OrderID, TotalPrice
FROM orders
ORDER BY TotalPrice DESC
LIMIT 5
"""
print(pd.read_sql(query3, conn))
```

```python
# STEP 7: GROUP BY + SUM (Aggregation)
# Total sales for each product
print("\n===== TOTAL SALES BY PRODUCT =====")
query4 = """
SELECT Product, SUM(TotalPrice) as Total_Sales
FROM orders
GROUP BY Product
ORDER BY Total_Sales DESC
"""
print(pd.read_sql(query4, conn))
```

```python
#STEP8: COUNT (Total number of records)
# Count total orders
print("\n===== TOTAL NUMBER OF ORDERS =====")
query5 = """
SELECT COUNT(*) AS Total_Orders
FROM orders
"""
print(pd.read_sql(query5, conn))
```

```python
#STEP9:AVG (Average calculation)
# Average order value
print("\n===== AVERAGE ORDER VALUE =====")
query6 = """
SELECT AVG(TotalPrice) AS Avg_Order_Value
FROM orders
"""
print(pd.read_sql(query6, conn))
```

```python
# STEP10:SUM (Total calculation)
# Total revenue of all orders
print("\n===== TOTAL REVENUE =====")
query7 = """
SELECT SUM(TotalPrice) AS Total_Revenue
FROM orders
"""
print(pd.read_sql(query7, conn))
```

```python
# STEP11:GROUP BY + COUNT
# Number of orders by payment method
print("\n===== ORDERS BY PAYMENT METHOD =====")
query8 = """
SELECT PaymentMethod, COUNT(*) AS Total_Orders
FROM orders
GROUP BY PaymentMethod
ORDER BY Total_Orders DESC
"""
print(pd.read_sql(query8, conn))
```

```python
# STEP12:GROUP BY + AVG
# Average order value by referral source
print("\n===== AVG ORDER VALUE BY REFERRAL SOURCE =====")
query9 = """
SELECT ReferralSource, AVG(TotalPrice) AS Avg_Value
FROM orders
GROUP BY ReferralSource
ORDER BY Avg_Value DESC
"""
print(pd.read_sql(query9, conn))
```

```python
# STEP13:FINAL INSIGHTS (Simple Explanation)
print("\n===== FINAL INSIGHTS =====")
print("1. High-value orders were identified using WHERE condition.")
print("2. Top products were found using GROUP BY and SUM.")
print("3. Total revenue and average order value were calculated.")
print("4. Payment methods were analyzed using GROUP BY.")
print("5. Referral sources show differences in average order value.")
```
