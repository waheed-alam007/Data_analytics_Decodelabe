```python
import pandas as pd
```

```python
df = pd.read_excel("/content/Dataset for Data Analytics.xlsx")
```

```python
df.head()
```

```python
df.info()
```

```python
# check the missing values
print("Missing Values in Each Column:")
df.isnull().sum()
```

```python
# fill the missing values
# This fills ALL missing values in the entire DataFrame with 0.
# Applies to every column (numeric, text, etc.)
# inplace=True means it directly updates df (no need to reassign)
df.fillna(0, inplace=True)
```

```python
#  This fills missing values (NaN) only in the CouponCode column.
# If a row has no coupon (NaN), it becomes: "No Coupon"
#  Keeps the column meaningful (since it's text data)
df['CouponCode'] = df['CouponCode'].fillna("No Coupon")
```

```python
# now i have to remove the missing values
df.isnull().sum()
```

```python
# now i  have to check the duplicate rows
df.duplicated().sum()
```

```python
# in this if you want to show the message plus the how many rows empty then  use this both above line of code judt dhoe the nmbr of empty rows and this below line also give the message and nmbr of row
print("Duplicate Rows Found:", df.duplicated().sum())
```

```python
# now remove the duplicate rows
df = df.drop_duplicates()
```

```python
# Check Duplicate Order IDs
df = df.drop_duplicates(subset='OrderID')
```

```python
# now standardize the date format
df['Date'] = pd.to_datetime(df['Date']).dt.strftime('%Y-%m-%d')
# verify date format
print(df['Date'].head())
```

```python
# now  i have to clean the txt cloumns Remove Extra Spaces
text_columns = ['Product', 'PaymentMethod', 'OrderStatus', 'ReferralSource']

for col in text_columns:
    df[col] = df[col].str.strip()
```

```python
print("FINAL VALIDATION REPORT")
print("-" * 30)

print("Missing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:", df.duplicated().sum())

print("\nDuplicate Order IDs:", df['OrderID'].duplicated().sum())
```
