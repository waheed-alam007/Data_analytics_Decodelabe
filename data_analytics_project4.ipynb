```python
# Import necessary libraries for data analysis and visualization
import pandas as pd              # For data handling
import matplotlib.pyplot as plt  # For basic plotting
import seaborn as sns            # For advanced visualization

# Set a clean style for graphs
sns.set(style="whitegrid")
```

```python
# Read the Excel file into a DataFrame
df = pd.read_excel("/content/Cleaned_Dataset.xlsx")

# Display first 5 rows to understand the data
df.head()
```

```python
# Show dataset structure (columns, data types, non-null values)
print(df.info())

# Convert 'Date' column into datetime format (important for time analysis)
df['Date'] = pd.to_datetime(df['Date'])

# Create a new column 'Year' from Date for trend analysis
df['Year'] = df['Date'].dt.year
```

```python
# Group data by Product and calculate total sales
product_sales = df.groupby('Product')['TotalPrice'].sum().sort_values(ascending=False)

# Create a bar chart
plt.figure(figsize=(8,5))
product_sales.plot(kind='bar')

# Add title and labels
plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Total Sales")

# Display the chart
plt.show()
```

```python
# Group data by Year and calculate total sales
year_sales = df.groupby('Year')['TotalPrice'].sum()

# Create a line chart
plt.figure()
year_sales.plot(kind='line', marker='o')

# Add title and labels
plt.title("Sales Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Total Sales")

# Display the chart
plt.show()
```

```python
# Count how many times each payment method is used
payment_counts = df['PaymentMethod'].value_counts()

# Create a bar chart
plt.figure()
payment_counts.plot(kind='bar')

# Add title and labels
plt.title("Payment Method Usage")
plt.xlabel("Payment Method")
plt.ylabel("Count")

# Display the chart
plt.show()
```

```python
# Count different order statuses (Delivered, Cancelled, etc.)
status_counts = df['OrderStatus'].value_counts()

# Create a bar chart
plt.figure()
status_counts.plot(kind='bar')

# Add title and labels
plt.title("Order Status Distribution")
plt.xlabel("Order Status")
plt.ylabel("Count")

# Display the chart
plt.show()
```

```python
# Create a scatter plot to see relationship between Quantity and Total Price
plt.figure()

sns.scatterplot(x='Quantity', y='TotalPrice', data=df)

# Add title
plt.title("Quantity vs Total Price")

# Display the chart
plt.show()
```

```python
# Select top 3 products based on total sales
top_products = product_sales.head(3)

# Create a pie chart
plt.figure()
top_products.plot(kind='pie', autopct='%1.1f%%')

# Add title and remove y-label for cleaner look
plt.title("Top 3 Product Contribution")
plt.ylabel("")

# Display the chart
plt.show()
```

```python
# Create a more professional-looking bar plot using seaborn
plt.figure(figsize=(8,5))

sns.barplot(x='Product', y='TotalPrice', data=df)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add title
plt.title("Professional Product Sales View")

# Display the chart
plt.show()
```
